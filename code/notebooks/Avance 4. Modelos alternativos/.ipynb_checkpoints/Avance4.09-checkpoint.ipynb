La serie de tiempos presentada en el grfico muestra una variabilidad significativa en el conteo de personas por da, con picos y cadas notables que indican fluctuaciones en la cantidad de personas capturadas.

Este comportamiento sugiere que puede haber factores externos o eventos que afectan la presencia de personas en las capturas, resultando en una serie de tiempo con alta variabilidad.

```python
conteo_por_dia = pd.DataFrame(grupos_dic['D']['sum'].value_counts()).reset_index()

# Definimos los rangos
bins = [0, 50, 100, 150, 200, 250, float('inf')]
labels = ['0-50', '51-100', '101-150', '151-200', '201-250', 'Mayor a 250']

# Creamos una nueva columna 'range' que indica el rango al que pertenece cada 'sum'
conteo_por_dia['range'] = pd.cut(conteo_por_dia['sum'], bins=bins, labels=labels, right=False)

# Agrupamos por la nueva columna 'range' y sumamos los 'count'
conteo_diario_por_rango = conteo_por_dia.groupby('range', observed=False)['count'].sum().reset_index()

conteo_diario_por_rango
```

```python
# Crear un grfico de barras
plt.figure(figsize=(5, 3))
plt.bar(conteo_diario_por_rango['range'], conteo_diario_por_rango['count'], color='skyblue', edgecolor='black', align='center')

# Aadir ttulos y etiquetas
plt.title('Conteo de Personas por dia')
plt.xlabel('Nmero de Personas')
plt.ylabel('Dias')
plt.xticks(range(len(conteo_diario_por_rango)))  # Asegura que el eje x tenga marcas de 0 a 10
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gcf().autofmt_xdate()

# Mostrar el grfico
plt.show()
```

# Ingenieria de caracteristicas

La ingeniera de caractersticas en series de tiempo es un paso crucial en el anlisis y modelado de datos temporales. Se centra en la transformacin de datos crudos en formatos que revelen ms informacin y sean ms adecuados para algoritmos de aprendizaje automtico. Renault et al.(2023) destacan la importancia de la ingeniera de caractersticas automtica para la clasificacin de series de tiempo, demostrando que los mtodos basados en caractersticas pueden ser tan precisos como los algoritmos de vanguardia. Adems, el uso de la Transformada Rpida de Fourier (FFT) como herramienta de ingeniera de caractersticas ha demostrado mejorar la precisin y eficiencia de los modelos de pronstico de series de tiempo.

*Referencias*
> - Galn-Sales, F.J., Reina-Jimnez , P., Carranza-Garca, M., Luna-Romera, J.M. (2023). An Approach to Enhance Time Series Forecasting by Fast Fourier Transform. In: Garca Bringas, P., et al. 18th International Conference on Soft Computing Models in Industrial and Environmental Applications (SOCO 2023). SOCO 2023. Lecture Notes in Networks and Systems, vol 749. Springer, Cham. https://doi.org/10.1007/978-3-031-42529-5_25 

> - Renault, A., Bondu, A., Lemaire, V., & Gay, D. (2023, June). Automatic Feature Engineering for Time Series Classification: Evaluation and Discussion. In 2023 International Joint Conference on Neural Networks (IJCNN) (pp. 1-10). IEEE. https://doi.org/10.1109/IJCNN54540.2023.10191074 


```python
from sklearn.linear_model import LinearRegression

from librerias_integrador.features.datetime_feat import create_date_time_features
```

En este conjunto de datos hemos creado una caracterstica de tendencia utilizando un modelo de regresin lineal, lo que permite capturar la direccin y velocidad del cambio en tus datos a lo largo del tiempo. 

```python
grupos_dic['D']['timestamp'] = pd.to_datetime(grupos_dic['D'].index)
grupos_dic['D']['trend'] = np.arange(len(grupos_dic['D']))

# Ajustar un modelo de regresin lineal para la tendencia
X = grupos_dic['D'][['trend']]  # Caracterstica de tendencia
y = grupos_dic['D']['sum']  # Variable objetivo
model = LinearRegression()
model.fit(X, y)

# Crear caractersticas de tendencia
grupos_dic['D']['trend_feature'] = model.predict(X)
```

Adems, hemos extrado caractersticas temporales como el ao, mes, da y da de la semana, as como un indicador binario para los fines de semana. Estas caractersticas pueden ayudar a los modelos a identificar y aprender patrones estacionales y tendencias a lo largo del tiempo, lo que es esencial para tareas como la prediccin y clasificacin en series de tiempo.

```python
daily_with_features = create_date_time_features(grupos_dic['D'], 'timestamp')[['sum','trend','trend_feature','year','month','day','weekday','is_weekend']]
daily_with_features.head()
```

El resultado de la ingeniera de caractersticas muestra una combinacin de datos numricos y categricos que reflejan tanto la tendencia subyacente como los patrones temporales especficos de los datos. La inclusin de estas caractersticas enriquecidas puede mejorar significativamente la capacidad de los modelos de aprendizaje automtico para hacer predicciones precisas y proporcionar insights valiosos a partir de los datos de series de tiempo.

# Modelos alternativos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
```

```python
# Diccionario para almacenar las mtricas de todos los modelos
model_metrics = {}

# Modelos a evaluar
models = {
    'Linear': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Poly': make_pipeline(PolynomialFeatures(degree=2), LinearRegression()),
    'DT': DecisionTreeRegressor(min_samples_split=3),
    'MLP': MLPRegressor(random_state=random_state, max_iter=20000)  # Red neuronal simple
}
```

La seleccin de modelos en machine learning debe equilibrar la precisin, la complejidad y la interpretabilidad. Por ello escogimos varios modelos alternativos para evaluar:

- Linear Regression: Es el punto de partida para la regresin debido a su simplicidad y eficiencia en el clculo. Es fcil de interpretar pero asume una relacin lineal entre las variables.
- Ridge y Lasso Regression: Son extensiones de la regresin lineal que incluyen regularizacin para prevenir el sobreajuste. Ridge es til cuando la mayora de las caractersticas son relevantes, mientras que Lasso puede seleccionar caractersticas importantes, descartando las menos significativas.
- Polynomial Regression: Captura relaciones no lineales al introducir trminos polinomiales. Aunque puede modelar curvas ms complejas, tambin es susceptible al sobreajuste.
- Decision Tree Regressor (DT): Puede capturar relaciones no lineales y es altamente interpretable. Sin embargo, es propenso al sobreajuste, especialmente si el rbol es muy profundo.
- MLPRegressor (MLP): Es un tipo de red neuronal que puede modelar interacciones complejas entre caractersticas. Ofrece una alta capacidad predictiva pero a costa de la interpretabilidad y un mayor costo computacional.

Referencias:
>-James, G., Witten, D., Hastie, T., & Tibshirani, R. (2013). An Introduction to Statistical Learning. Springer.
>
>-LeCun, Y., Bengio, Y., & Hinton, G. (2015). Deep learning. Nature, 521(7553), 436-444.

```python
# Entrenamiento y evaluacin de modelos
for name, model in models.items():
    model.fit(X_train, y_train)
    print('\n'*3,name)
    metrics = plot_model_predictions(model, X_train, X_test, y_train, y_test, return_metrics=True)
    model_metrics[name] = metrics
```

```python
model_metrics['dummy'] = metricas_por_modelo['dummy']
# Convertir el diccionario en un DataFrame
metricas_por_modelo_df = pd.DataFrame.from_dict({(i,j): model_metrics[i][j]
                             for i in model_metrics.keys()
                             for j in model_metrics[i].keys()},
                            orient='index')

# Renombrar los ndices
metricas_por_modelo_df.index = pd.MultiIndex.from_tuples(metricas_por_modelo_df.index, names=["modelo", "dataset"])
metricas_por_modelo_df
```

```python
def plot_metrics(metricas_por_modelo_df, titles):
    """
    Genera un grid de grficos scatter plot para las mtricas de modelos de machine learning.

    Parmetros:
    - metricas_por_modelo_df (DataFrame): Un DataFrame con las mtricas de los modelos.
    - titles (list): Una lista de ttulos de mtricas a graficar.

    Salida:
    - Cuatro grficos scatter plot en un grid de 2x2.
    """
    # Crear figuras y ejes para un grid de 2x2
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 5))

    # Iterar sobre cada mtrica y crear un grfico para cada una
    for i, metric in enumerate(titles):
        ax = axes[i//2, i%2]
        # Filtrar los datos de entrenamiento y prueba para la mtrica actual
        train_data = metricas_por_modelo_df.xs('train', level='dataset')[metric]
        test_data = metricas_por_modelo_df.xs('test', level='dataset')[metric]
        
        # Obtener los nombres de los modelos para el eje x
        x_labels = train_data.index
        
        # Crear scatter plot para los datos de entrenamiento y prueba
        ax.scatter(x_labels, train_data, marker='o', label='Train')
        ax.scatter(x_labels, test_data, marker='x', label='Test')
        
        ax.plot(x_labels, train_data, color='blue', linestyle='-', linewidth=0.5)
        ax.plot(x_labels, test_data, color='orange', linestyle='-', linewidth=0.5)
        
        # Establecer el ttulo y las etiquetas
        ax.set_title(metric)
        ax.set_ylabel(metric)
        
        # Aadir leyenda
        ax.legend()

    # Ajustar el layout y mostrar el grfico
    plt.tight_layout()
    plt.show()
```

```python
# Ttulos de los grficos
metricas = ['rmse','mae','mape','r2']
plot_metrics(metricas_por_modelo_df, metricas)
```

# Hiperparametros

```python
import scipy.stats as stats
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.metrics import make_scorer

from librerias_integrador.models.metrics import neg_root_mean_squared_error
from librerias_integrador.models.plots import plot_cv_results
```

El uso de RandomizedSearchCV ofrece varias ventajas en la optimizacin de hiperparmetros para modelos de machine learning:

- Eficiencia: Al seleccionar una muestra aleatoria del espacio de hiperparmetros, reduce significativamente el tiempo computacional en comparacin con una bsqueda exhaustiva, especialmente cuando el espacio es grande.
- Exploracin: Permite una exploracin ms diversa del espacio de hiperparmetros, lo que puede llevar al descubrimiento de combinaciones inesperadas que funcionan bien.
- Flexibilidad: Es menos susceptible a quedarse atascado en ptimos locales gracias a su naturaleza aleatoria.
- Escalabilidad: Se adapta bien a espacios de alta dimensin y es efectivo incluso cuando algunos hiperparmetros son ms importantes que otros.

```python
cv_n_splits = 3

param_distributions = {
    'Linear':{'fit_intercept':[True, False]},
    'Ridge': {'alpha': np.logspace(-4, 4, 279)},
    'Lasso': {'alpha': np.logspace(-4, 4, 279)},
    'DT': {'max_depth': np.arange(1, 17), 'min_samples_leaf': np.arange(3, 19)},
    'MLP': {'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)], 'activation': ['tanh', 'relu'], 'alpha':np.logspace(-4, 0.1, 279).tolist()},
    'Poly': {'polynomialfeatures__degree': np.arange(1, 5), 'linearregression__fit_intercept': [True, False]}
}

cv = KFold(n_splits=cv_n_splits, shuffle=True, random_state=random_state)
```

```python
total_combinations = {model: np.prod([len(values) for values in params.values()]) for model, params in param_distributions.items()}
print(f"Total de combinaciones de parmetros por modelo: {total_combinations}")
```

```python
rmse_scorer = make_scorer(neg_root_mean_squared_error, greater_is_better=True)
```

```python
min_iter = 107
best_regressors = {}

# Realizar RandomizedSearchCV para cada modelo
for model_name, model in models.items():
    print('\n'*3,model_name)
    parameters = param_distributions[model_name]
    n_iter = min(min_iter, sum(len(values) for values in parameters.items()))
    search = RandomizedSearchCV(model, 
                                parameters, 
                                n_iter=n_iter, cv=cv, 
                                random_state=random_state,
                               return_train_score=True,
                               scoring=rmse_scorer)
    search.fit(X, y)
    
    # Almacenar el mejor regresor en el diccionario
    best_regressors[model_name] = search.best_estimator_
    
    # Usar la funcin plot_cv_results para visualizar los resultados
    plot_cv_results(search.cv_results_, cv)
    
    # Imprimir los mejores parmetros encontrados
    print(f'Mejores parmetros para {model_name}: {search.best_params_}')
    print('\n'*3)
```

```python
model_metrics_tuned = {}
# Despus del bucle for, puedes acceder a los mejores modelos as:
for model_name, best_model in best_regressors.items():
    print('\n'*3,model_name)
    # Usar la funcin plot_model_predictions para visualizar las predicciones del mejor modelo
    model_metrics_tuned[model_name] = plot_model_predictions(best_model, X_train, X_test, y_train, y_test, return_metrics=True)
```

```python
model_metrics_tuned['dummy'] = metricas_por_modelo['dummy']

# Convertir el diccionario en un DataFrame
metricas_por_modelo_tuned_df = pd.DataFrame.from_dict({(i,j): model_metrics_tuned[i][j]
                             for i in model_metrics_tuned.keys()
                             for j in model_metrics_tuned[i].keys()},
                            orient='index')

# Renombrar los ndices
metricas_por_modelo_tuned_df.index = pd.MultiIndex.from_tuples(metricas_por_modelo_tuned_df.index, names=["modelo", "dataset"])
metricas_por_modelo_tuned_df
```

# Analisis de resultados

Al analizar el rendimiento de los modelos de regresin, observamos que tanto el DecisionTreeRegressor (DT) como el MLPRegressor (MLP) superan a los dems modelos en trminos de las mtricas de rendimiento clave: RMSE, MAE, MAPE, y R^2.

El DT muestra una mejora significativa en las mtricas de entrenamiento en comparacin con las de prueba, lo que sugiere un posible sobreentrenamiento. Este fenmeno se caracteriza por un modelo que aprende los datos de entrenamiento tan bien que pierde generalizacin, resultando en un rendimiento inferior en datos no vistos. Adicionado al comportamiento de bloques observado en el grfico de correlacin de el conteo real vs el predicho, identificndose 3 escalones y no un comportamiento lineal.

Por otro lado, el MLP muestra un equilibrio ms consistente entre las mtricas de entrenamiento y prueba, lo que indica una mejor generalizacin a nuevos datos. Aunque el DT tiene un R^2 ligeramente superior en la prueba, la consistencia del MLP en ambas fases sugiere que es ms confiable para hacer predicciones en situaciones del mundo real.

Adems, el MLP muestra que existe un potencial para aplicar aprendizaje profundo, que es una rama avanzada de machine learning capaz de modelar interacciones complejas y aprender representaciones de alto nivel de los datos. Esto lo hace particularmente valioso para tareas donde las relaciones entre las caractersticas no son lineales o bien definidas.

En trminos de trade-offs, aunque el MLP puede ser ms difcil de interpretar que el DT, la importancia de la interpretabilidad debe sopesarse contra la precisin y la generalizacin. En muchas aplicaciones industriales y comerciales, la capacidad de generalizar y predecir con precisin puede ser ms valiosa que la explicabilidad del modelo.

En resumen, el MLPRegressor parece ser el modelo ms robusto y confiable para este conjunto de datos, equilibrando el rendimiento en los datos de entrenamiento y prueba y ofreciendo ventajas en el contexto del aprendizaje profundo. Aunque el DT puede ser ms fcil de explicar, el sobreentrenamiento y la menor generalizacin sugieren que el MLP es una mejor eleccin para aplicaciones prcticas donde la precisin y la capacidad de generalizacin son primordiales.
