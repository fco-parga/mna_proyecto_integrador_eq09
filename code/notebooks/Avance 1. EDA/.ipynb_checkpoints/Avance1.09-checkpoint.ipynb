```python
import matplotlib.dates as mdates
```

```python
def plot_df_counts(df, figsize=(13, 3), grouped=False, agg2plot=None, intervalo=''):
    """
    Genera un grfico de lneas del conteo de personas a lo largo del tiempo.

    Parmetros:
    df (DataFrame): DataFrame de pandas que contiene los datos a graficar.
    figsize (tuple, opcional): Tamao de la figura del grfico.
    grouped (bool, opcional): Indica si los datos estn agrupados por alguna mtrica.
    agg2plot (str, opcional): Nombre de la columna agregada a graficar si los datos estn agrupados.
    intervalo (str, opcional): Intervalo de tiempo de los datos para incluir en el ttulo.

    Devuelve:
    None: Esta funcin no devuelve nada, solo muestra el grfico.
    """

    # Configura el tamao de la figura del grfico
    plt.figure(figsize=figsize)
    
    # Grafica los datos agrupados o el conteo de personas por captura
    if grouped and agg2plot:
        plt.plot(df.index, df[agg2plot], '-o', markersize=2, label='Conteo')  
        metrica = agg2plot
    else:
        plt.plot(df.index, df['people'], '-o', markersize=2, label='Conteo')   
        plt.plot(df.index, np.zeros(df.shape[0]), 'o', markersize=2, color='darkgreen', label='Capturas')  
        metrica = 'por captura'
    
    # Configura el formato de las etiquetas del eje x para fechas
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator())
    
    # Ajusta la orientacin de las etiquetas de fecha
    plt.gcf().autofmt_xdate()
    
    # Establece las etiquetas y el ttulo del grfico
    plt.xlabel('Fecha')
    plt.ylabel('Nmero de personas')
    titulo = f'Conteo de personas ({metrica}) {intervalo}'
    plt.title(titulo)
    
    # Activa la grilla para mejor visualizacin
    plt.grid(True)
    # Muestra la leyenda del grfico
    plt.legend()
    
    # Muestra el grfico
    plt.show()
```

```python
plot_df_counts(raw_df)
```

En este grfico se muestra el conteo de personas por captura cada 30 minutos aproximadamente. 

La **<span style="color:blue">lnea azul</span>** representa el conteo instantneo de la captura, mientras que los **<span style="color:green">puntos verdes</span>** representan la hora de la captura.

# Agragaciones temporales

Las agregaciones temporales en series de tiempo ofrecen varias ventajas clave:

- Reduccin del Ruido: Al agrupar datos en intervalos ms grandes, se suavizan las fluctuaciones aleatorias, lo que facilita la identificacin de tendencias y patrones.
- Eficiencia Computacional: Al disminuir la granularidad de los datos, se reduce el volumen de datos a procesar, lo que puede mejorar la velocidad de clculo y anlisis.
- Mejora en la Visualizacin: Las agregaciones pueden hacer que los grficos de series de tiempo sean ms legibles y fciles de interpretar, al presentar la informacin de manera ms clara y concisa.
- Anlisis de Tendencias: Facilita el anlisis de tendencias a largo plazo al eliminar variaciones estacionales o cclicas de corto plazo.

*Referencias*
> - Pea, D. (2010). Anlisis de series temporales. Madrid: Alianza Editorial

```python
grupos_dic = {}
agrupaciones = ['2h', '4h', 'D', 'W']

for intervalo in agrupaciones:

    grupos_dic[intervalo] = raw_df.resample(intervalo).agg(
        mean=('people', 'mean'),
        sum=('people', 'sum')
    )
```

```python
for intervalo in agrupaciones:
    for agg in ['mean','sum']:
        plot_df_counts(grupos_dic[intervalo], grouped=True, agg2plot=agg, intervalo=intervalo)
```

```python
conteo_por_dia = pd.DataFrame(grupos_dic['D']['sum'].value_counts()).reset_index()

# Definimos los rangos
bins = [0, 50, 100, 150, 200, 250, float('inf')]
labels = ['0-50', '51-100', '101-150', '151-200', '201-250', 'Mayor a 250']

# Creamos una nueva columna 'range' que indica el rango al que pertenece cada 'sum'
conteo_por_dia['range'] = pd.cut(conteo_por_dia['sum'], bins=bins, labels=labels, right=False)

# Agrupamos por la nueva columna 'range' y sumamos los 'count'
conteo_diario_por_rango = conteo_por_dia.groupby('range', observed=False)['count'].sum().reset_index()

conteo_diario_por_rango
```

```python
# Crear un grfico de barras
plt.figure(figsize=(5, 3))
plt.bar(conteo_diario_por_rango['range'], conteo_diario_por_rango['count'], color='skyblue', edgecolor='black', align='center')

# Aadir ttulos y etiquetas
plt.title('Conteo de Personas por dia')
plt.xlabel('Nmero de Personas')
plt.ylabel('Dias')
plt.xticks(range(len(conteo_diario_por_rango)))  # Asegura que el eje x tenga marcas de 0 a 10
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gcf().autofmt_xdate()

# Mostrar el grfico
plt.show()
```

## Funciones auxiliares

```python
def get_annotated_img_objects(img, detector_output, score_threshold=0.50, onlyhumans=True):
    """
    Anota objetos detectados en una imagen basndose en la salida de un detector.

    Parmetros:
    img (np.array): La imagen original en la que se detectarn los objetos.
    detector_output (dict): Un diccionario con las salidas del detector, incluyendo clases, cajas y puntuaciones.
    score_threshold (float, opcional): El umbral de puntuacin para considerar una deteccin vlida.
    onlyhumans (bool, opcional): Si es True, solo se anotarn detecciones humanas.

    Devuelve:
    tuple: Una tupla conteniendo la imagen anotada, el nmero de objetos detectados,
           los ndices de las detecciones, las cajas detectadas y los tipos de objetos.
    """

    # Extrae la informacin relevante del output del detector
    class_ids = detector_output.get("detection_classes")[0]
    boxes = detector_output.get("detection_boxes")[0]
    scores = detector_output.get("detection_scores")[0]
    num_detections = int(detector_output.get("num_detections")[0])

    # Convierte la imagen de entrada a un array de numpy y prepara las variables
    image = np.array(img)
    objects = 0
    y_scale, x_scale, _ = image.shape
    ids = []
    object_type = []
    detected_boxes = {}

    # Decide qu objetos se van a anotar
    if onlyhumans:
        objects2plot = [1]  # Solo humanos
    else:
        objects2plot = list(set(class_ids.astype(int)))  # Todos los objetos detectados

    # Itera sobre todas las detecciones
    for i in range(num_detections):
        # Verifica si la deteccin cumple con el umbral y si es un objeto a anotar
        if scores[i] >= score_threshold and int(class_ids[i]) in objects2plot:
            box = boxes[i]
            y_min, x_min, y_max, x_max = box

            # Dibuja un rectngulo alrededor del objeto detectado
            cv.rectangle(image, (int(x_min*x_scale), int(y_min*y_scale)), 
                         (int(x_max*x_scale), int(y_max*y_scale)), (0, 255, 0), 5)
            objects += 1
            ids.append(i)
            object_type.append(int(class_ids[i]))
            detected_boxes[i] = [(int(x_min*x_scale), int(y_min*y_scale)), 
                                 (int(x_max*x_scale), int(y_max*y_scale))]

    # Devuelve la imagen anotada junto con informacin sobre las detecciones
    return image, objects, ids, detected_boxes, object_type
```

```python
def detect_persons(img_list, detector, score_threshold=0.85, plot=True, size=(13,13)):
    """
    Detecta personas en una lista de imgenes utilizando un detector proporcionado.

    Parmetros:
    img_list (list): Lista de rutas de archivos de imgenes para procesar.
    detector (modelo): Modelo de deteccin de objetos para identificar personas.
    score_threshold (float, opcional): Umbral de puntuacin para filtrar detecciones.
    plot (bool, opcional): Si es True, muestra las imgenes con las detecciones.
    size (tuple, opcional): Tamao de la imagen a mostrar.

    Devuelve:
    dict: Diccionario con imgenes etiquetadas y sus cajas de deteccin.
    dict: Diccionario con cajas de deteccin por imagen.
    """

    # Inicializa diccionarios para almacenar resultados
    images_labeled = {}
    boxes_per_image = {}

    # Procesa cada imagen en la lista
    for filename in tqdm(img_list):
        # Lee la imagen y la convierte al espacio de color RGB
        img = cv.imread(filename)
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

        # Obtiene la salida del detector para la imagen actual
        detector_output = detector(tf.expand_dims(img, axis=0))
        # Anota la imagen con las detecciones de personas
        img_detect, objects, ids, detected_boxes, object_type = get_annotated_img_objects(img, detector_output, score_threshold=score_threshold)
        # Almacena la imagen anotada y las detecciones en el diccionario
        images_labeled[filename.split("\\")[-1]] = {'img': img_detect, 'boxes':detected_boxes, 'object':object_type}
        boxes_per_image[filename.split("\\")[-1]] = detected_boxes

    # Si se debe mostrar, itera sobre las imgenes etiquetadas y las muestra
    if plot:
        for name, image in images_labeled.items():
            mostrar_imagen(image['img'], cmap='', title=name, size=size)
            
    # Devuelve los diccionarios con las imgenes etiquetadas y las cajas de deteccin
    return images_labeled, boxes_per_image
```

```python
def plot_images_in_grid(images_labeled, columns=3, cmap='gray', size=(23, 23)):
    """
    Muestra una cuadrcula de imgenes con anotaciones.

    Parmetros:
    images_labeled (dict): Un diccionario con nombres de imgenes y sus datos correspondientes.
    columns (int, opcional): Nmero de columnas en la cuadrcula.
    cmap (str, opcional): Mapa de colores para mostrar las imgenes.
    size (tuple, opcional): Tamao de la figura en pulgadas.

    Devuelve:
    None: Esta funcin no devuelve nada, solo muestra las imgenes en una cuadrcula.
    """

    # Calcula el nmero total de imgenes
    total_images = len(images_labeled)
    
    # Calcula el nmero de filas necesarias
    rows = total_images // columns + (total_images % columns > 0)
    
    # Configura el tamao de la figura basado en el nmero de filas y columnas
    plt.figure(figsize=(size[0], size[1] * rows))
    
    # Itera a travs de las imgenes y sus nombres
    for index, (name, image) in enumerate(images_labeled.items()):
        # Crea un subplot para cada imagen
        plt.subplot(rows, columns, index + 1)
        # Muestra la imagen con o sin un mapa de colores
        if cmap == "":
            plt.imshow(image['img'])
        else:
            plt.imshow(image['img'], cmap)
        # Cuenta el nmero de objetos detectados
        n_counts = len(image['object'])
        # Define el ttulo con el nombre y el nmero de personas detectadas
        title = f'{name}\nPersonas detectadas: {n_counts}'
        plt.title(title, fontsize=10)
        plt.axis('off')  # Oculta los ejes
    
    plt.tight_layout()  # Ajusta la disposicin de los subplots
    plt.show()  # Muestra la cuadrcula con todas las imgenes
```
